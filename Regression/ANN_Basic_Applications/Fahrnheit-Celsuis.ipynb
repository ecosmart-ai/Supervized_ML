```
import tensorflow  as tf
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#import dataset 
from google.colab import drive
drive.mount('/content/drive')
```

```
df = pd.read_csv('/content/drive/My Drive/Data/Celsius+to+Fahrenheit.csv')
df.info()
```

```
sns.scatterplot(x='Celsius',y='Fahrenheit',data=df)
```

```
X_train = df['Celsius']
y_train = df['Fahrenheit']
```

```
model =tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1 ,  input_shape=[1]))
```

```
model.compile(optimizer=tf.keras.optimizers.Adam(0.5),loss='mean_squared_error')
```

```
epochs_hist = model.fit(X_train,y_train,epochs=200)
```

```
plt.plot(epochs_hist.history['loss'])
```

```
model.get_weights()
```

```
T_c = [0]
T_f = model.predict(T_c)
print('value :' , T_f)
```
